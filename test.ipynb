```python
import numpy as np
from collections import Counter

def predict_with_model(input_data, model, scaler=None, needs_scaling=False):
    input_array = np.asarray(input_data).reshape(1, -1)
    if needs_scaling and scaler is not None:
        input_array = scaler.transform(input_array)
    prediction = model.predict(input_array)
    return int(prediction[0])

# Input features 
input_data = (37, 1, 0, 2, 130, 250, 0, 1, 187, 0, 3.5, 0)
print("Input data: ",input_data)
models = {
    "Logistic Regression": (log_reg, True),
    "KNN": (knn, True),
    "XGBoost": (xgb_clf, False),
    "Random Forest": (rf_clf, False),
    "Gradient Boosting": (gb_clf, False),
    "Extra Trees": (et_clf, False),
    "AdaBoost": (ada_clf, False),
    "LightGBM": (lgbm_clf, False),
}

predictions = []

for name, (model, needs_scaling) in models.items():
    pred = predict_with_model(input_data, model, scaler=scaler, needs_scaling=needs_scaling)
    #print(f"{name} Prediction: {prediction}")
    predictions.append(pred)

count = Counter(predictions)
majority_pred, majority_count = count.most_common(1)[0]

if majority_count > len(models) / 2:
    result = "Person has the disease" if majority_pred == 1 else "Person does NOT have the disease"
    print(f"{majority_pred} --> {result}")
else:
    print("No clear majority prediction among models.")
```
